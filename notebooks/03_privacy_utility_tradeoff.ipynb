# 03 — Privacy / utility trade-off

Plot final accuracy vs spent epsilon for the DP runs, and add the SecAgg + full-pipeline
endpoints. This is the figure that anchors the diploma's results chapter.

````python
import json, pathlib
import matplotlib.pyplot as plt

def final_accuracy(history_path):
    h = json.loads(pathlib.Path(history_path).read_text())
    series = h.get('metrics_distributed', {}).get('accuracy', [])
    if not series:
        return None
    return series[-1][1]

runs = {
    'vanilla': 'results/vanilla_fedavg/history.json',
    'dp': 'results/federated_dp/history.json',
    'secagg': 'results/federated_secagg/history.json',
    'full': 'results/federated_full_pipeline/history.json',
}

for name, path in runs.items():
    if pathlib.Path(path).exists():
        print(f'{name:>8s}: final acc = {final_accuracy(path)}')
````
