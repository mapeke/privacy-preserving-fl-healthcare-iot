# 02 — Baseline vs federated learning

Compare the centralized baseline against vanilla FedAvg on the same data partition.
Reads results JSON files written by the experiment scripts.

```python
import json, pathlib
import matplotlib.pyplot as plt

central = json.loads(pathlib.Path('results/centralized/centralized_results.json').read_text())
fed = json.loads(pathlib.Path('results/vanilla_fedavg/history.json').read_text())
print('centralized test accuracy:', central['test_accuracy'])
```

```python
rounds, losses = zip(*fed['losses_distributed'])
plt.plot(rounds, losses, marker='o', label='FedAvg distributed loss')
plt.axhline(central['test_loss'], linestyle='--', label='Centralized test loss')
plt.xlabel('Round')
plt.ylabel('Loss')
plt.legend()
plt.title('Centralized vs federated training')
```
